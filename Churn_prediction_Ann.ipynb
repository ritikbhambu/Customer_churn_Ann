```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
```

```python
dataset
```

```python
#dividing into dependent and independent featured
```

```python
x = dataset.iloc[:,3:13]
y = dataset.iloc[:,13]
```

```python
x
```

```python
y
```

```python
# feature engineering
geography  = pd.get_dummies(x['Geography'],drop_first=True)
gender = pd.get_dummies(x['Gender'],drop_first=True)
```

```python
geography
```

```python
# add this with df
x=x.drop(['Geography','Gender'],axis = 1)
```

```python
pd.concat([x,geography,gender],axis=1)
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# feature scaling is required for - Ann,linear reg,logistic reg, kknn,k means
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# creating ann
# using tensorflow
```

```python
from tensorflow.keras.models import Sequential # for  forward and bakward propagation
from tensorflow.keras.layers import Dense # used to create input,hidden,op layer
from tensorflow.keras.layers import LeakyReLU,PReLU,ELU,ReLU # activation function
from tensorflow.keras.layers import Dropout  # for dead nodes
```

```python
classifier = Sequential()
```

```python
# input layer
classifier.add(Dense(units=11,activation='relu')) # inputs = 11
```

```python
# first hidden layer
classifier.add(Dense(units=7,activation='relu'))
```

```python
#second hidden layer
classifier.add(Dense(units=6,activation='relu'))
```

```python
# adding op layer
classifier.add(Dense(units=1,activation='sigmoid')) # for binary = signmoid
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history = classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
# here we have to perfirn early stopping
# early stopping is when accuracy is not increasing in epochs
```

```python
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor = "val_loss",
    min_delta = 0.0001,
    patience = 20,
    verbose = 1,
    mode = "auto",
    baseline = None,
    restore_best_weights = False,
)
```

```python
# prediction
y_pred = classifier.predict(x_test)
y_pred = (y_pred > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn  import metrics
score = metrics.accuracy_score(y_test,y_pred)
score
```
